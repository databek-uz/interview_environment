```python
from pyspark.sql import SparkSession
import pandas as pd
```

```python

spark = SparkSession.builder \
    .appName("InterviewTest") \
    .master("local[*]") \
    .getOrCreate()
```

```python
pdf = pd.DataFrame({
    "name": ["Ali", "Vali"],
    "age": [30, 25]
})

df = spark.createDataFrame(pdf)
df.show()

spark.stop()
```
